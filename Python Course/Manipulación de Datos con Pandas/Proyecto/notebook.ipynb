## ¿Qué escuelas de la ciudad de Nueva York tienen los mejores resultados en matemáticas?

Los mejores resultados en matemáticas son al menos el 80% de la puntuación máxima posible de 800 * en matemáticas.
Guarde sus resultados en un `DataFrame` de `pandas` llamado `best_math_schools`, incluidas las columnas `"school_name"`y `"average_math"`, ordenados `"average_math"`en orden descendente.

```python
filter_sort_schools_math = schools[schools["average_math"] >= 800 * 0.8].sort_values("average_math", ascending = False)
```

```python
best_math_schools = filter_sort_schools_math[["school_name", "average_math"]]
```

## ¿Qué distrito tiene la desviación estándar más grande en el puntaje combinado del SAT?

Guarde sus resultados como un `DataFrame` de `pandas` llamado `largest_std_dev`.

1- El `DataFrame` debe contener una fila, con:

* `"borough"`- el nombre del distrito de Nueva York con la desviación estándar más grande de `"total_SAT"`.
* `"num_schools"`- el número de escuelas del municipio.
* `"average_SAT"`- la media de `"total_SAT"`.
* `"std_SAT"`- la desviación estándar de `"total_SAT"`.

2- Redondea todos los valores numéricos a dos decimales.

```python
SAT_schools = schools[["school_name", "borough", "total_SAT"]].sort_values("total_SAT", ascending = False)
```

```python
SAT_schools.head()
```

```python
std_dev_by_borough = SAT_schools.groupby("borough")['total_SAT'].std().reset_index()
```

```python
std_dev_by_borough.head()
```

```python
max_std_borough_df = std_dev_by_borough[std_dev_by_borough["total_SAT"] == std_dev_by_borough["total_SAT"].max()]
```

```python
num_schools_df = SAT_schools[SAT_schools["borough"] == "Manhattan"].count()
```

```python
max_std_borough = max_std_borough_df["borough"]
```

```python
num_schools = num_schools_df["borough"]
```

### Media de `total_SAT`

```python
mean_by_borough = SAT_schools.groupby("borough")["total_SAT"].mean().reset_index()
```

```python
mean_of_total_SAT_df = mean_by_borough[mean_by_borough["borough"] == "Manhattan"]
```

```python
mean_of_total_SAT = mean_of_total_SAT_df["total_SAT"]
```

### STD de `total_SAT`

```python
std_by_borough = SAT_schools.groupby("borough")["total_SAT"].std().reset_index()
```

```python
std_of_total_SAT_df = std_by_borough[std_by_borough["borough"] == "Manhattan"]
```

```python
std_of_total_SAT = std_of_total_SAT_df["total_SAT"]
```

### Crear DataFrame

```python
largest_std_dev = pd.DataFrame()
```

```python
largest_std_dev["borough"] =  max_std_borough
largest_std_dev["num_schools"] = num_schools
largest_std_dev["average_SAT"] = round(mean_of_total_SAT, 2)
largest_std_dev["std_SAT"] = round(std_of_total_SAT, 2)
```
